```python
%matplotlib inline

import torch
import torch.optim as optim
from torchvision import transforms, models
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
vgg = models.vgg19(pretrained=True).features

for param in vgg.parameters():
    param.requires_grad_(False)
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
vgg.to(device)
```

```python
def load_image(img_path, max_size=400, shape=None):
    image = Image.open(img_path).convert('RGB')
    if max(image.size) > max_size:
        size = max_size
    else:
        size = max(image.size)

    if shape is not None:
        size = shape

    in_transform = transforms.Compose([
        transforms.Resize(size),
        transforms.ToTensor(),
        transforms.Normalize((0.5,0.5,0.5),
                             (0.5,0.5,0.5))
    ])
    image = in_transform(image).unsqueeze(0)
    return image
```

```python
content = load_image('images/City.jpg').to(device)
style = load_image('images/StarryNight.jpg').to(device)
```

```python
def im_convert(tensor):
    image = tensor.cpu().clone().detach().numpy()
    image = image.squeeze()
    image = image.transpose(1, 2, 0) # move channels: Color x w x h -> w x h x color
    image = image * np.array((0.5, 0.5, 0.5)) + np.array((0.5, 0.5, 0.5))
    image = image.clip(0, 1)
    return image
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(im_convert(content))
ax1.axis('off')
ax2.imshow(im_convert(style))
ax2.axis('off')
```

```python
content = load_image('images/City.jpg').to(device)
style = load_image('images/StarryNight.jpg', shape=content.shape[-2:]).to(device)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))
ax1.imshow(im_convert(content))
ax1.axis('off')
ax2.imshow(im_convert(style))
ax2.axis('off')
```

```python
def get_features(image, model):
    layers = {
        '0': 'conv1_1',
        '5': 'conv2_1',
        '10': 'conv3_1',
        '19': 'conv4_1',
        '21': 'conv4_2', # content extraction
        '28': 'conv5_1'
    }
    
    features = {}
    
    for name, layer in model._modules.items():
        image = layer(image) # Re-process image through all the layers in the model
        if name in layers:
            features[layers[name]] = image # Save specific layered image

    return features

content_features = get_features(content, vgg)
style_features = get_features(style, vgg)
```

```python
def gram_matrix(tensor):
    _, d, h, w = tensor.size()
    tensor = tensor.view(d, h*w)
    gram = torch.mm(tensor, tensor.t())
    return gram
```

```python
style_grams = { layer: gram_matrix(style_features[layer]) for layer in style_features }
```

```python
style_weights = {
    'conv1_1': 1.,
    'conv2_1': 0.75,
    'conv3_1': 0.2,
    'conv4_1': 0.2,
    'conv5_1': 0.2
}

# nice style: low alpha/beta ratio
content_weight = 1 # alpha
style_weight = 1e6 # beta
```

```python
target = content.clone().requires_grad_(True).to(device)
```

```python
# Optimization
show_every = 300 # show target image every 300 iterations to see progress
optimizer = optim.Adam([target], lr=0.003)
steps = 2100 # balance style output and processing time, more better accuracy but it takes more time
# steps = 10000 # great results for animation

# for data visualization, create a video
height, width, channels = im_convert(target).shape
image_array = np.empty((300, height, width, channels))
capture_frame = steps / 300   # capture a frame every 7 steps
counter = 0
```

```python
for ii in range(1, steps+1):
    target_features = get_features(target, vgg)
    content_loss = torch.mean( (target_features['conv4_2'] - content_features['conv4_2']) ** 2 )
    style_loss = 0
    for layer in style_weights:
        target_feature = target_features[layer]
        target_gram = gram_matrix(target_feature)
        style_gram = style_grams[layer]
        layer_style_loss = style_weights[layer] * torch.mean( (target_gram - style_gram)**2 )
        _, d, h, w = target_feature.shape
        style_loss += layer_style_loss / (d * h * w)

    total_loss = content_loss * content_weight + style_loss * style_weight
    
    optimizer.zero_grad()
    total_loss.backward()
    optimizer.step()
    
    if ii % show_every == 0:
        print('Total loss: ', total_loss.item())
        print('Iteration: ', ii)
        plt.imshow(im_convert(target))
        plt.axis('off')
        plt.show()
    
    if ii % capture_frame == 0:
        image_array[counter] = im_convert(target)
        counter += 1
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,10))
ax1.imshow(im_convert(content))
ax1.axis('off')
ax2.imshow(im_convert(style))
ax2.axis('off')
ax3.imshow(im_convert(target))
ax3.axis('off')
```

```python
# conda install -c https://conda.binstar.org/menpo opencv
import cv2

frame_height, frame_width, _ = im_convert(target).shape
vid = cv2.VideoWriter('output.mp4', cv2.VideoWriter_fourcc(*'XVID'), 30, (frame_width, frame_height))

for i in range(0, 300):
    img = image_array[i]
    img = img * 255
    img = np.array(img, dtype=np.uint8)
    cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # opencv: images as BGR
    vid.write(img)

vid.release() # write, flush, close video file

```
